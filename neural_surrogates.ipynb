```python
import os
from time import time
from tqdm import tqdm

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import pde
```

```python
features = np.load('features.npy')
targets = np.load('targets.npy')

m_min, m_max = features.min(), features.max()
d_min, d_max = targets.min(), targets.max()

mm = (features - m_min) / (m_max - m_min)
dd = (targets - d_min) / (d_max - d_min)
print('Features: {} | Targets: {}'.format(mm.shape, dd.shape))

fig, axs = plt.subplots(5, 2, figsize=(7, 10), sharex=True, sharey=True)
for i in range(5):
    ax1, ax2 = axs[i]
    k = i * 80
    ax1.imshow(mm[k], cmap='jet', interpolation='none')
    ax2.imshow(dd[k], cmap='jet', interpolation='none')
    ax1.set_ylabel('R{}'.format(k))
    [a.set_title(['Model','Data'][j], color=['b','r'][j]) for j, a in enumerate([ax1, ax2])] if i == 0 else None
plt.tight_layout()
plt.show()
```
